# tf.keras.layers.Embedding #把文本向量化，把单词变成索引(把文本映射为一个密集向量)

```python
import re
```

```python
token = re.compile('[A-Za-z]+|[!?,.()]')
```

```python
def reg_text(text):
    new_text = token.findall(text)
    new_text = [word.lower() for word in new_text]
    return new_text
```

```python
data['text'] = data.text.apply(reg_text)
```

```python
data
```

```python
word_set = set()
for text in data.text:
    for word in text:
        word_set.add(word)
```

```python
len(word_set)
```

```python
word_list = list(word_set)
```

```python
word_list
```

```python
word_list.index('stressed')
```

```python
word_index = dict((word, word_list.index(word) + 1) for word in word_list)
```

```python
word_index
```

```python
#获取text每个单词的索引，如果搜索不到就返回0
data_ok = data.text.apply(lambda x: [word_index.get(word, 0) for word in x])
```

```python
len(data_ok.iloc[0]), len(data_ok.iloc[1])
```

```python
maxlen = max(len(x) for x in data_ok)
```

```python
maxlen
```

```python
max_word = len(word_set) + 1
max_word
```

```python
data_ok
```

```python
#填充sequence方法
data_ok = keras.preprocessing.sequence.pad_sequences(data_ok.values, maxlen = maxlen)
```

```python
data_ok.shape
```

```python
data.review.values
```

```python
model = keras.Sequential()
```

```python
model.add(layers.Embedding(max_word, 50, input_length=maxlen))
```

```python
model.add(layers.LSTM(64))
```

```python
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['acc'])
```

```python
model.fit(data_ok, 
          data.review.values, 
          epochs = 10, 
          batch_size = 128, 
          validation_split = 0.2)
```
